## Markowitz mean-variance Model Portfolio

```python
"""
依照馬可維茲的比率去分
這邊的 rho 我取1/2
"""

port_rsd=pd.DataFrame( columns = ['Return','Std'])
Daily_return= pd.DataFrame( columns = ["MMM","AXP","AAPL","BA","CAT","CVX","CSCO","KO","GS","HD","HON","IBM","INTC","JNJ",
                               "JPM","MCD","MRK","MSFT","NKE","PG","CRM","TRV","UNH","VZ","V","wba","WMT","DIS"])
list2=[]
list3=[]
for ticker in tickers :
    r1=-1
    list1=[]
    return_port=0
    return_port_tmp=0
    for i in range(len(df2.loc[:,ticker])):
        return_port=(df2.loc[i,ticker]-r1)/r1
        if(r1 == -1): return_port =0
        list1.append(return_port) ##dataframe 不能append資料，只能用list或array裝
        r1=df2.loc[i,ticker]
    Daily_return[ticker]=list1
    sum_return=sum(list1)/len(df2.loc[:,ticker])
    list2.append(sum_return)
    r1=-1
    list1=[]
    for i in range(len(df2.loc[:,ticker])):
        return_port_tmp=(df2.loc[i,ticker]-r1)/r1
        return_port_tmp=pow((return_port_tmp-return_port),2)
        if(r1 == -1): return_port_tmp =0
        list1.append(return_port_tmp)
        r1=df2.loc[i,ticker]
    sum_dev=pow(sum(list1)/(len(df2.loc[:,ticker])-1),1/2)
    list3.append(sum_dev)
Daily_return=Daily_return.drop(0)
port_rsd['Return']=list2
port_rsd['Std']=list3 
```
